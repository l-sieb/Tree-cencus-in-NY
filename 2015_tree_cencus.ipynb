```python
mask = ((tree_subset['status']=='Dead') | (tree_subset['status']=='Stump'))
```

```python
tree_subset.loc[mask] = tree_subset.loc[mask].fillna('Not Applicable')
```

```python
tree_subset[tree_subset['status']=='Stump']
```

```python
# Let's look at the remaining missing values
tree_subset.isna().sum()
```

We can look at how to fill the NaN values

```python
tree_subset[tree_subset['health'].isna()]
```

```python
tree_subset['health'].fillna('None', inplace=True)
```

```python
tree_subset[tree_subset['problems'].isna()]
```

```python
tree_subset['problems'].value_counts()
#The NaN values can be replaced by None
```

```python
tree_subset['problems'].fillna('None', inplace = True)
```

```python
tree_subset[tree_subset['spc_latin'].isna()]
```

```python
tree_subset['spc_latin'].value_counts()
```

```python
tree_subset['spc_latin'].fillna('No Observation', inplace = True)
tree_subset['spc_common'].fillna('No Observation', inplace = True)
```

```python
tree_subset[tree_subset['sidewalk'].isna()]
```

```python
tree_subset['sidewalk'].value_counts()
```

```python
tree_subset['sidewalk'].fillna('NoDamage', inplace = True)
```

```python
tree_subset[tree_subset['guards'].isna()]
```

```python
tree_subset['guards'].value_counts()
```

```python
tree_subset['guards'].fillna('None', inplace = True)
```

```python
# Did all the replacement work :
tree_subset.isna().sum()
```
